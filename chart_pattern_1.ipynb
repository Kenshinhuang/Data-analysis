```python
from pandas_datareader import data
from datetime import datetime
from datetime import date
import pandas as pd
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go
from datetime import timedelta
import numpy as np
```

#### Enter the stock's ticker symbol, then press Enter on the keyboard.

```python
ticker=input("The stock's ticker symbol: ")
```

```python
df = data.DataReader(f'{ticker}', 'yahoo', '2016-01-01')
df=df.reset_index()
all_time_high=df["Close"].max()
date1=df[df["Close"]==all_time_high][["Date"]]
lowest_since=df[df["Date"]>=date1["Date"].to_list()[0]]["Close"].min()
second_highest=df[df["Date"]>date1["Date"].to_list()[0]]["Close"].max()
date3=df[df["Close"]==second_highest][["Date"]]
date4=df[df["Close"]==lowest_since][["Date"]]
date2=date1["Date"]-timedelta(days=25)
df1=df[df["Date"]>=date2.to_list()[0]]

date11=datetime.date(date1["Date"].to_list()[0])
today=date.today()

if date11==today:
    print (f"{ticker} already broke out from all barriers and reached its all time high today.")
    fig = go.Figure(data=[go.Candlestick(x=df['Date'],
                open=df['Open'],
                high=df['High'],
                low=df['Low'],
                close=df['Close'])])
    fig.update_layout(xaxis_rangeslider_visible=False,title=f"{ticker}")
    fig.show()
else:
    print (f"{ticker} is performing under a resistance since falling from its all-time high.")
    date_11=df[df["Date"]>=date1["Date"].to_list()[0]]
    date_31=df[df["Date"]>=date3["Date"].to_list()[0]]
    date_41=df[df["Date"]>=date4["Date"].to_list()[0]]
    fig = go.Figure(data=[go.Candlestick(x=df1['Date'],
                open=df1['Open'],
                high=df1['High'],
                low=df1['Low'],
                close=df1['Close'])])
    fig.add_trace(go.Scatter(
    x=date_11["Date"],
    y=np.repeat(all_time_high,len(date_11)),
    mode="lines+text",
    text=["all time high"],
    textposition="top right",line_color="purple"))
    fig.add_trace(go.Scatter(
    x=date_31["Date"],
    y=np.repeat(second_highest,len(date_31)),
    mode="lines+text",
    text=["peak since falling from all time high"],
    textposition="bottom right",line_color="blue"))
    fig.add_trace(go.Scatter(
    x=date_41["Date"],
    y=np.repeat(lowest_since,len(date_41)),
    mode="lines+text",
    text=["lowest since falling from all time high"],
    textposition="bottom right",line_color="black"))
    fig.update_layout(font_size=15,showlegend=False,xaxis_rangeslider_visible=False,title=f"{ticker}")
    fig.show()
```
